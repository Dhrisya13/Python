```python
#create a Bank account with members account number,name,type of account and balance.
#write constructor and methods to deposit at the bank and withdraw an amount from the bank.
class BankAccount:
    def __init__(self,name,account_no,account_type,balance):
        self.name=name
        self.account_no=account_no
        self.account_type=account_type
        self.balance=balance
    def deposit(self,dep):
        print("initial balance:",self.balance)
        if dep>0:
            self.balance+=dep
            print("the deposited amount is:",dep)
            print("The current balance is:",self.balance)
    def withdraw(self,amount):
        #self.amount=int(input("withdraw amount:"))
        print("initial balance:",self.balance)
        if amount>self.balance:
            print("insufficient balance")
        elif amount<=0:
            print("invalid withdrawal amount")
        else:
            self.balance-=amount
            print(f"{amount} withdrawn successfully")
            print("Current balance:", self.balance)
    def display(self):
        print("customer account name:",self.name)
        print("account number:",self.account_no)
        print("account type:",self.account_type)
        print("account balance:",self.balance)
name=input("Your name:")
account_no=int(input("Account number:"))
account_type=input("Your account type(Savings/current):")
balance=0
obj=BankAccount(name,account_no,account_type,balance)
while True:
    print("\n1.Deposit \n2.Withdraw \n3.Display \n4.Exit")
    n=input("Enter your choice:")
    if n=='1':
        obj.deposit(float(input("Depositing Amount:")))
    elif n=='2':
        obj.withdraw(float(input("Withdrawal Amount:")))
    elif n=='3':
        obj.display()
    elif n=='4':
        print("Exiting")
        break
    else:
        print("invalid choice")
```
