```python
# create a class Rectangle with private attributes length and width.Overload '<' operator
# to compare the area of 2 rectangles.

class Rectangle:
    def __init__(self,l,w):
        self.__l=l
        self.__w=w
        
    def area(self):
        return self.__l * self.__w
    
    def __lt__(self,other):
        return self.area() < other.area()
    
l1=int(input("Enter the length of first rectangle:"))
w1=int(input("Enter the width of first rectangle:"))
l2=int(input("Enter the length of second rectangle:"))
w2=int(input("Enter the width of second rectangle:"))
r1=Rectangle(l1,w1)
r2=Rectangle(l2,w2)
print("Area of 1st rectangle:",r1.area())
print("Area of 2nd rectangle:",r2.area())
if r1 < r2:
    print("First rectangle is smaller than the second.")
else:
    print("First rectangle is not smaller than the second.")
```

```python
#Create a class Time with private attributes hour,minute,and second.Overload '+' operator to find sum of 2 time.
class Time:
    def __init__(self,hour,minute,second):
        self.hour=hour
        self.minute=minute
        self.second=second
    def display(self):
        print(f"Hour: {self.hour}  Min: {self.minute}  Sec: {self.second}")
    def __add__(self,other):
        h=self.hour + other.hour
        m=self.minute + other.minute
        s=self.second + other.second
        
        if s>=60:
            m+=s//60 #floor division-->to get the quotient value
            s=s%60 #modulus-->to get the remainder
        if m>=60:
            h+=m//60
            m=m%60
            return Time(h, m, s)
        
h1=int(input("Enter the hour1:"))
m1=int(input("Enter the minute1:"))
s1=int(input("Enter the seconds1:"))
h2=int(input("Enter the hour2:"))
m2=int(input("Enter the minute2:"))
s2=int(input("Enter the seconds2:"))

t1=Time(h1,m1,s1)
t2=Time(h2,m2,s2)
t3=t1+t2
t1.display()
t2.display()
t3.display()
```
