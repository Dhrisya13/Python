```python
# WPP using regular expression to extract all email addresses from a given text.
# Sample text:text="Contact us at info@abc.com or Support@xyz.org for more details"

import re
text=text="Contact us at info@abc.com or Support@xyz.org for more details."
pattern=r"[a-zA-Z-0-9.+]+@[a-zA-Z0-9.-]+\.[a-zA-Z]{2,}"
email=re.findall(pattern,text)
print("Email address found",email)
```
